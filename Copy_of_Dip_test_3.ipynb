Q2. Now, display an appropriate visualization of the proportions of color components in the image.[2marks]

```python
#write your code here
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
image = Image.open('/content/drive/MyDrive/dataset/woman-headscarf-exotic-beautiful.jpg')


image_array = np.array(image)
r, g, b = image_array[:, :, 0], image_array[:, :, 1], image_array[:, :, 2]


bins = 256
 plt.figure(figsize=(8, 5))

plt.hist(r.ravel(), bins=bins, color='red', alpha=0.6, label='Red')
plt.hist(g.ravel(), bins=bins, color='green', alpha=0.6, label='Green')
plt.hist(b.ravel(), bins=bins, color='blue', alpha=0.6, label='Blue')

plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.title('Color Component Distribution')
plt.legend()
plt.show()
```

```python
#Write your code here
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image


image = Image.open('/content/drive/MyDrive/dataset/woman-headscarf-exotic-beautiful.jpg')
image = np.array(image) / 255.0


mean = 0
std_dev = 1
noise = np.random.normal(mean, std_dev, image.shape)
noisy_image = image + noise


noisy_image = np.clip(noisy_image, 0, 1) * 255
noisy_image = noisy_image.astype(np.uint8)


denoised_image = cv2.GaussianBlur(noisy_image, (5,5), 1)


fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(noisy_image)
ax[0].set_title("Noisy Image")
ax[0].axis("off")

ax[1].imshow(denoised_image)
ax[1].set_title("Denoised Image (Gaussian Filter)")
ax[1].axis("off")

plt.show()
```

Q5. Now apply any appropriate technique to detect the edges of the filtered image.
[6 points]

```python
#Write your Answer here
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

image = Image.open('/content/drive/MyDrive/dataset/woman-headscarf-exotic-beautiful.jpg')
image = np.array(image) / 255.0

# Apply Gaussian noise
mean = 0
std_dev = 1
noise = np.random.normal(mean, std_dev, image.shape)
noisy_image = image + noise


noisy_image = np.clip(noisy_image, 0, 1) * 255
noisy_image = noisy_image.astype(np.uint8)

# Apply Gaussian filter to remove noise
denoised_image = cv2.GaussianBlur(noisy_image, (5,5), 1)

# Convert to grayscale for edge detection
gray_image = cv2.cvtColor(denoised_image, cv2.COLOR_RGB2GRAY)


edges = cv2.Canny(gray_image, threshold1=50, threshold2=150)
# Display results
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].imshow(denoised_image)
ax[0].set_title("Denoised Image")
ax[0].axis("off")

ax[1].imshow(edges, cmap='gray')
ax[1].set_title("Edge Detection (Canny)")
ax[1].axis("off")

plt.show()
```
